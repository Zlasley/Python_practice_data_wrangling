Data wrangling, library importing, and loading in data with a URL.

```python
# Imports
# Normally done at the top of a notebook to show the user what will be needed.
import pandas as pd
import numpy as np

# Assiging the cell phone data URL to a value
cell_phones_url = 'https://raw.githubusercontent.com/LambdaSchool/data-science-practice-datasets/main/unit_1/Cell__Phones/cell_phones.csv'

# Loading in the dataframe
cell_phones = pd.read_csv(cell_phones_url)

# displaying the top 5 rows
cell_phones.head()
```

```python
# Displaying the last 5 rows from the dataframe
cell_phones.tail()
```

A 2nd time with a different dataframe

```python
population_url = 'https://raw.githubusercontent.com/LambdaSchool/data-science-practice-datasets/main/unit_1/Population/population.csv'

# Load the dataframe

population = pd.read_csv(population_url)

# print the first 5 records
population.head()
```

```python
assert isinstance(population, pd.DataFrame)
```

A 3rd time with a dataframe consisting geo codes.

```python
geo_codes_url = 'https://raw.githubusercontent.com/LambdaSchool/data-science-practice-datasets/main/unit_1/GEO_codes/geo_country_codes.csv'

# Load the dataframe

geo_codes = pd.read_csv(geo_codes_url)

# print the first 5 records
geo_codes.head()
```

Checking for missing values in the three dataframes.

```python
# Check for missing data in each of the DataFrames

cell_phones_missing = cell_phones.isnull().sum().sum()

population_missing = population.isnull().sum().sum()

geo_codes_missing = geo_codes.isnull().sum().sum()

print(f"Cell phone DF has {cell_phones_missing} null values")
print(f"Population DF has {population_missing} null values")
print(f"Geo codes DF has {geo_codes_missing} null values")
```

Merging dataframes into one.

```python
# Merging the cell_phones and population dataframes

cell_phone_population = (pd.merge(cell_phones, population,))

cell_phone_population
```

Merge the cell_phone_population and geo_codes DataFrames

```python
# Merging the cell_phone_population and geo_codes dataframes
# While only including the country and geo columns from geo_codes

geo_cell_phone_population = pd.merge(cell_phone_population, geo_codes[['geo', 'country']], how='inner')

geo_cell_phone_population
```

Lets calculate the number cell phones per person and place the results in its own column.

```python
# Here we divided the value of phones by population and created a new column called "phones_per_person"
geo_cell_phone_population['phones_per_person'] = cell_phone_population['cell_phones_total'] / cell_phone_population['population_total']
geo_cell_phone_population
```

Lets dive in and see the number of cell phones per person in the US in 2017

```python
# Determine the number of cell phones per person in the US in 2017
US_2017 = geo_cell_phone_population.loc[8455]

# View the DataFrame
US_2017
```

We continue on to look at the summary statistics and find the mean value for "phones_per_person"

```python
# Statistics summary
geo_cell_phone_population.describe()
```

```python
# Mean value for "phones_per_person"
mean_phones = round(geo_cell_phone_population.describe().loc['mean'].loc['phones_per_person'],2)

mean_phones
```

Now we are going to create a new dataframe called df2017 which will include only records from from geo_cell_phone_population that ocurred in 2017.


```python
df2017 = geo_cell_phone_population.loc[geo_cell_phone_population['time'] == 2017]

df2017.head()
```

Finally lets find the five countries with the most cell phones per person in 2017.

```python
# Sorted the values in descending order and named them df2017_top
df2017_top = df2017.sort_values(['phones_per_person'], ascending=[False]).head()

# Displaying the final result
df2017_top
```
